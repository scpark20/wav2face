```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.load('infer08.14-1.npy')
data1 = data1.mean(axis=0)

data2 = np.load('infer08.14-2.npy')
data2 = data2.mean(axis=0)

data3 = np.load('infer08.14-3.npy')
data3 = data3.mean(axis=0)

print(data1.shape)
```

```python
names = [(17, 'JawOpen'), (20, 'MouthPucker'), (23, 'MouthSmileLeft'), (37, 'MouthLowerDownLeft')]

for name in names:
    plt.figure(figsize=[18, 2])
    plt.title(str(name[0]) + ' : ' + name[1])
    plt.plot(data1[name[0]], label='Reg.', linewidth=2, color='red')
    plt.plot(data2[name[0]], label='DVAE', linewidth=2, color='blue')
    plt.plot(data3[name[0]], label='Flow', linewidth=2, color='green')
    plt.grid()
    plt.legend()
    plt.show()
```
